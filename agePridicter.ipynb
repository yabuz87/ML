```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("new.csv")
df
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[['length']],df.age)
```

```python
reg.predict([[150]])
```

```python
%matplotlib inline
plt.scatter(df.length,df.age, color='red', marker='*')
plt.xlabel("X-axis(length)")
plt.ylabel("Y-axis(Age)")
plt.plot(df['length'], reg.predict(df[['length']]), color='blue')
plt.show()
```
