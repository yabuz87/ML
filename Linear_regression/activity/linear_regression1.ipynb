```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn import linear_model
```

```python
df=pd.read_csv("canada_per_capita_income.csv")
df
```

```python
%matplotlib inline
plt.scatter(df.year, df[r'per capita income (US$)'], marker="*")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
print(df.columns)
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[['year']],df[r'per capita income (US$)'])
```

```python
reg.predict([[2020]])
```

```python
# Import necessary libraries
%matplotlib inline
import matplotlib.pyplot as plt

# Scatter plot for data
plt.scatter(df['year'], df[r'per capita income (US$)'], color='red', marker='*')  

# Plot regression line (assuming 'reg' is already fitted with appropriate data)
plt.plot(df['year'], reg.predict(df[['year']]), color='blue')  

# Add axis labels
plt.xlabel('Year', fontsize=20)
plt.ylabel('Per Capita Income (US$)', fontsize=20)

# Show the plot
plt.show()
```
